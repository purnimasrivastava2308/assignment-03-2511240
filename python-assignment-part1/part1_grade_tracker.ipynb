## Task 1 — Data Parsing & Profile Cleaning

Initializing a list of dictionary containing students information

```python
raw_students = [
    {"name": "  ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma",       "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair  ",    "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA",       "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ",    "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]
```

Updating the raw_students list:
1. Removed the leading/trailing white space and converted into title case
2. Typecasted roll no into integer data type
3. Typecasted and converted marks_str into integer data type list

```python
for i in range(len(raw_students)):
    raw_students[i]["name"] = raw_students[i]["name"].strip().title()       #Assigning the value of dictionary key 'name' after removing leading/trailing whitespace and converting it in title case
    raw_students[i]['roll'] = int(raw_students[i]['roll'])                  #converting the string value of key 'roll' into int
    raw_students[i]['marks_str'] = raw_students[i]['marks_str'].split(', ') #converting the string value of 'marks_str' into string list
    raw_students[i]['marks'] = list(map(int, raw_students[i]['marks_str'])) #Converting string list into int and assigning it a new key 'marks'
    del raw_students[i]['marks_str']                                        #Deleting the key-value pair of dictionary with key name 'marks_str'
```

Checking name from raw_students dictionary:
1. Valid, if it contains only alphabet
2. Invalid, otherwise

```python
for dict_ele in raw_students:
    valid = False
    name_list = dict_ele["name"].split(" ") #Extracting value of name key from dictionary and spliting into list
    for name in name_list:                  #Iterating over name from name_list list
        if name.isalpha():                  
            valid = True
            
    if valid:
        print(dict_ele["name"],"✓ Valid name")
    else:
        print(dict_ele["name"],"✗ Invalid name")
```

Formatted Profile Card of each student

```python
for i in range(len(raw_students)):
    print(f"="*30)
    print(f"Student : {raw_students[i]["name"]}")
    print(f"Roll no : {raw_students[i]["roll"]}")
    print(f"Marks   : {raw_students[i]["marks"]}")
    print(f"="*30)
    print()
```

Printing the name in upper and lower case for roll no. 103

```python
print(raw_students[2]["name"].upper())  #Accessing the list element which contains a dictionary with roll no '103' and converting it in lower case
print(raw_students[2]["name"].lower())
```

## Task 2 — Marks Analysis Using Loops & Conditionals

Initilization

```python
student_name = "Ayesha Sharma"
subjects     = ["Math", "Physics", "CS", "English", "Chemistry"]
marks        = [88, 72, 95, 60, 78]
```

Print each subject alongside its marks and a grade label using f string

```python
print(f"{'Subject':<20} {'marks':<12} {'grade':<12}")
grade = ""
for i in range(len(marks)):
    if marks[i] > 90:
        grade = "A+"
    elif marks[i] > 80:
        grade = "A"
    elif marks[i] > 70:
        grade = "B"
    elif marks[i] > 60:
        grade = "C"
    else:
        grade = "F"
    
    print(f"{subjects[i]:<20} {marks[i]:<12} {grade}")
```

Printing the sum, average, maximum, minimum marks respectively

```python
print("Total marks", sum(marks))
print("Total marks", sum(marks)/len(marks))
print("Total marks", max(marks))
print("Total marks", min(marks))
```

Adding a new subject with marks 

```python
mark = 0
count = 0
while True:
    new_sub = input("Enter the new subject you want to add (write 'done' to exit)").strip().title()
    if new_sub.lower() == 'done':   #Checking the exit condition
        break

    while True: 
        try:                                                    #Handle the value error if mark is non-integer
            mark = int(input("Enter marks (0-100)").strip())    
            if mark >= 0 and mark <= 100: 
                subjects.append(new_sub)        #appending the subject if marks is integer and ranges between 0-100
                marks.append(mark) 
                count += 1                      #Counting number of subjects added
                break                           #Exit the inner while loop
            else:
                print("Warning: do not crash, and do not add invalid entries to the list")
                continue

        except ValueError:                      #Handling exception when user enters anything except digits
            print("Warning: do not crash, and do not add invalid entries to the list")

print(f"Number of new subject added: {count}")
print(marks)
new_avg = sum(marks)/len(marks)
print(f"New average {new_avg:.2f}")         #Showing average till 2 decimal place
```

# Task 3 — Class Performance Summary 

Initialization of class_data

```python
class_data = [
    ("Ayesha Sharma",  [88, 72, 95, 60, 78]),
    ("Rohit Verma",    [55, 68, 49, 72, 61]),
    ("Priya Nair",     [91, 85, 88, 94, 79]),
    ("Karan Mehta",    [40, 55, 38, 62, 50]),
    ("Sneha Pillai",   [75, 80, 70, 68, 85]),
]
```

Calculating student's average and their pass/fail result

```python
avg = []
result_status = []

for index in range(len(class_data)):            #Iteraing over class_data list
    value = round(sum(class_data[index][1])/len(class_data[index][1]), 2)   #Calculating average and rounding off to 2 decimal place
    avg.append(value)                           #Appending average marks per student
    if avg[index] >=60:                         #Checking their result status- pass/fail
        result_status.append('Pass')
    else:
        result_status.append('Fail')
```

Presenting the data in formatted form

```python
print(f"{'Subject':<20} | {'Average':<7} | {'Status':<8}")          #Displaying titles
print("-"*45)
for i in range(len(class_data)):
    print(f"{class_data[i][0]:<20} | {avg[i]:<7} | {result_status[i]:<8}")  #Displaying individual student's result
```

Showing count of students who passwd and failed respectively
Along with class topper name and average as well as
Class average

```python
p = result_status.count("Pass")
print(f"Number of student who passed: {p}")
print(f"Number of student who failed: {len(result_status)-p}")

i = avg.index(max(avg))
print(f"Class topper: {class_data[i][0]} {max(avg)}")
print(f"Class average: {sum(avg)/len(avg):.2f}")
```

# Task 4 — String Manipulation Utility

Initialized a string value in eassy variable

```python
essay = "  python is a versatile language. it supports object oriented, functional, and procedural programming. python is widely used in data science and machine learning.  "
```

Removing the leading/trailing whitespace and converted the string into title case

```python
clean_essay = essay.strip().title()
print(clean_essay)
```

Counting occurences of word 'python'. Later on replacing it with 'Python 🐍'
Splitting the original string on '. '

```python
count = clean_essay.lower().count("python")
print(f"Number of occurence of 'python': {count}")
print(clean_essay.lower().replace("python", "Python 🐍"))
split_list = clean_essay.split(". ")
print(split_list)
```

Printing each sentence on a new line with numbering

```python
for i in range(len(split_list)):
    if '.' not in split_list[i]:
        print(i+1, ". ", split_list[i], ".", sep="")
    else:
        print(i+1, ". ", split_list[i], sep="")
```
