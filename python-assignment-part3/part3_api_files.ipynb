Creating a text file and writing content in it

```python
import requests
from pathlib import Path

# Get the directory of the current script
try:
    base = Path(__file__).parent
except NameError:
    base = Path.cwd()   # Current Working Directory

file_path = base / "python_notes.txt"

with open(file_path, "w", encoding="utf-8") as f:
    f.write("Topic 1: Variables store data. Python is dynamically typed.\n")
    f.write("Topic 2: Lists are ordered and mutable.\n")
    f.write("Topic 3: Dictionaries store key-value pairs.\n")
    f.write("Topic 4: Loops automate repetitive tasks.\n")
    f.write("Topic 5: Exception handling prevents crashes.\n")

print("File written successfully.")
```

Appending 2 more lines in the file

```python
with open(file_path, "a", encoding="utf-8") as f:
    f.write("Topic 6: Tuples are immutable.\n")
    f.write("Topic 7: Set does not contain duplicate elements.\n")
    
print("Lines appended.")
```

Couting total number of lines in the file along with priting each lines numbered

```python
count = 0
with open(file_path, "r", encoding="utf-8") as f:
    for number, line in enumerate(f, start=1):
        print(f"{number}:", line)
        count += 1
```

```python
print("The total number of lines in the file:", count)
```

```python
keyword = input("Enter a word you want to search (type 'exit' if don't want to search)").strip().lower()
flag = 0
with open(file_path, "r", encoding="utf-8") as f:
    for line in f:
        if keyword in line.lower():
            print(line)
            flag = 1
    
if flag == 0:
    print("Oopsiee!! Cannot find your special word. Wanna append :) ")
```
