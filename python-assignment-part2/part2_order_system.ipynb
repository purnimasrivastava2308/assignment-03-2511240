## Task 2 — Cart Operations

Defined a function to add/update an item in the cart

```python
def add_item(cart):
    d = {'item':"", 'quantity':0, 'price':""}       #Defining an empty dictionary to add/update the item in the cart
    avail = False                                   #Supposing item is not present in the cart
    index = -1                                      #Supposing the index number of the item (user want to add and is already present in the cart) in the cart 

    item = input("Enter the item you want to add in the menu").strip().title()      #Add an item in the cart
    if item in menu.keys():                         #Checking if the item is present in the menu

        for i in range(len(cart)):                  #Checking if the item already exists in the cart
            if cart[i]['item'] == item:
                avail = True                        
                index = i

        if menu[item]['available']:             #Checks if the item is available in the menu
            quantity = int(input(f"Enter the quantity for your order item {item}"))

            if avail == False:          #Add the item in the cart if it doesnot exist earlier
                d['item'] = item
                d['quantity'] = quantity
                d['price'] = (menu[item]['price'] * quantity)
                cart.append(d)

            else:                       #Update the quantity and price of the element
                cart[index]['quantity'] += quantity
                cart[index]['price'] += (menu[item]['price']*quantity)

        else:
            print("Item is not available")

    else:
        print("Item doesn't exist in the menu")

    return cart                         #returning the updated cart
```

Updating cart items and print order summary

```python
cart = []

while True:
    user = input(f"Enter what operation you want to perform (add/delete) or 'done' to order)").strip().lower()
    if user == 'done':          #Exit case
        break
    elif user == 'add':         #Add/update item
        cart = add_item(cart)
    elif user == 'delete':      #Delete item
        cart = delete_item(cart)
    else:                       #Ensuring valid input is received from the customer
        print("Enter valid input")


print("============== Order Summary ==============")
sum = 0
for i in range(len(cart)):
    print(f"{cart[i]['item']:<20} x{cart[i]['quantity']:<10} ₹{cart[i]['price']:>7.2f}")
    sum += cart[i]['price']         #Calculating the total cart amount

print("-"*43)
print(f"{'Subtotal:':<32} ₹{sum:>7.2f}")
print(f"{'GST (5%):':<32} ₹{(0.05*sum):>7.2f}")
print(f"{'Total Payable:':<32} ₹{sum + (0.05*sum):>7.2f}")
print("===========================================")
```

## Task 4 — Daily Sales Log Analysis 

Calculating the total sales per day

```python
k = list(sales_log.keys())
sum = 0                         #Total sales in a day
s = []                          #List that contains total sales of each day

for i in range(len(k)):         
    for j in range(len(sales_log[k[i]])):
        sum += sales_log[k[i]][j]['total']      #Total sales per day
    s.append(sum)                               #Appending total sales per day in the list
```

Printing total revenue per day and the best selling day

```python
print("Total revenue per day")
for i in range(len(s)):
    print(f"{k[i]}: ₹{s[i]}")

print("\nThe best-selling day")

i = s.index(max(s))
print(f"{k[i]}: ₹{s[i]}")
```

```python
from collections import defaultdict

count = defaultdict(int)        #Initializing dictionary with default int value
for i in sales_log:             #Iterating over dates in the sales_log
    for j in sales_log[i]:      #Iterating over items list in the date key
        for k in j['items']:    #Iterating over each item from the item list
            count[k] += 1       #Counting occurence of each item
print(count)

most_ordered = max(count, key=count.get)        #Finding the element with maximum order 
print("Most ordered item:", most_ordered)
print("Number of orders:", count[most_ordered])
```
